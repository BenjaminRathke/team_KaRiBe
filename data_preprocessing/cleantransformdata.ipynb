```python
#Import Dependencies
import pandas as pd
```

```python
#import initial raw data
nominees_df = pd.read_csv('Song Of The Year Award.csv', encoding='cp1252')
songatt_df = pd.read_csv('songAttributes_1999-2019.csv')
grammysongs_df = pd.read_csv('the_grammy_awards.csv')
```

```python
#create a dataframe of nominated but losing songs
losers_df = nominees_df[(nominees_df['Winner'] == False) & (nominees_df['Year'] < 2019)]
losers_df = losers_df.rename(columns={'Song Name': 'song_name', 'Year': 'year', 'Artist': 'artist', 'Winner': 'winner'})
losers_df
```

```python
# create winners dataframe where nominee column is changed to song_name for concat/merge purposes,
# and filters out only data for song of the year and years 1999-2019.

grammysongs_df = grammysongs_df.rename(columns={"nominee": "song_name"})
winners_df = grammysongs_df[(grammysongs_df['category']=='Song Of The Year') & (grammysongs_df['year'] >= 1999) & (grammysongs_df['year'] < 2019)]
winners_df
```

```python
#concat dataframes of winners & losers to have full set of all nominated
winnerslosers_df = pd.concat([losers_df, winners_df], axis=0)
winnerslosers_df
```

```python
#update song attribute df column "Name" to "song_name"for merge purposes.
songatt_df = songatt_df.rename(columns={"Name": "song_name"})
```

```python
songdatadirty_df = winnerslosers_df.merge(songatt_df, how='left', on='song_name')

#SEE NOTES BELOW ABOUT THE SONGDATA.CSV.
#songdatadirty_df.to_csv('songdata.csv')

songdatadirty_df
```

```python
# Since there was no appropriately matched "artist" column to merge songatt_df and winnerslosers_df on,
# duplicates matched with incorrect song attribute data was manually removed from the songdata.csv file created above.
# The manually cleaned file is now imported here to coninue cleaning with python/pandas.

songdatacleaner_df = pd.read_csv('songdata.csv')
songdatacleaner_df
```

```python
#drop rows with no song attribute data
songdatacleaner_df = songdatacleaner_df.dropna(subset=['Valence'])
#drop rows that have text values or values that wouldn't logically impact outcome.
songdatacleaner_df = songdatacleaner_df.drop(['song_name', 'artist', 'year', 'title', 'published_at', 'updated_at', 'category', 'workers', 'img', 'Unnamed: 0', 'Album', 'Artist'], axis=1)
songdatacleaner_df
```

```python
#change boolean values to integers for machine learning purposes for final machine learning dataset.
songdatacleaned_df = songdatacleaner_df.convert_dtypes()
songdatacleaned_df["winner"] = songdatacleaner_df["winner"].astype(int)
songdatacleaned_df["Explicit"] = songdatacleaner_df["Explicit"].astype(int)

#export and display final machine learning dataset
songdatacleaned_df.to_csv('songdata_cleaned.csv', index=False)
songdatacleaned_df
```
